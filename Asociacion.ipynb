```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_excel("ComprasClean.xlsx", index_col = "Unnamed: 0")
df.head()
```

```python
infoProduct = (df.groupby(["id_Proveedor","Cod_Producto"])["Compra_Cantidad"].sum().unstack().reset_index().fillna(0).set_index("id_Proveedor")) 
infoProduct
```

```python
def binarizar(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
dfMatriz = infoProduct.applymap(binarizar)
dfMatriz
```

```python
frecuencia = apriori(dfMatriz, min_support = 0.07, use_colnames = True)
reglas = association_rules(frecuencia, metric= "lift", min_threshold = 1)
reglas
```
